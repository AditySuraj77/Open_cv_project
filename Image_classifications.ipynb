```python
import cv2
import cvzone
from cvzone.ClassificationModule import Classifier
```

```python
my_classifier = Classifier('keras_model.h5','labels.txt')
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    
    ret, frame = cap.read()
    
    if ret == True:
        predic = my_classifier.getPrediction(frame)
        
        cv2.imshow('Predictions',frame)
        
        k = cv2.waitKey(50)
        
        if k == 27:
            break
            
cap.release()
cv2.destroyAllWindows()
```
